```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.target
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
# features selection for model

X = df.iloc[:, :-1]
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
```

```python
treemodel = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train, y_train)
```

```python
plt.Figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)
plt.show()
```

```python
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```

```python
feature_importances = treemodel.feature_importances_

plt.barh(iris.feature_names, feature_importances)
plt.xlabel("Feature Importance")
plt.title("Feature Importance in Decision Tree Classifier")
plt.show()
```

So in post pruning i'm using simple max depth 2 limits to two levels , tree can only use at most 2-3 features remaining features get zero importance coz they're never used in the tree. 

```python
## preprunning

parameters = {
    "criterion": ["gini", "entropy", "log_loss"],
    "splitter": ["best", "random"],
    "max_depth": [1, 2, 3, 4, 5],
    "max_features": ["auto", "sqrt", "log2"],
}
```

```python
treemodel = DecisionTreeClassifier(max_depth=2)
cv = GridSearchCV(treemodel, param_grid=parameters, cv=5, scoring="accuracy")
# treemodel defining the model DecisionTreeClassifier , param_grid = assigning parameters which we initialized in key value we assign param_grid = parameters , cv = cross validation train data - validation data (divides the splitted data and takes those validation data find accuracy)
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
```

```python
score
```

```python
print(classification_report(y_pred, y_test))
```

```python
feature_importances = cv.best_estimator_.feature_importances_

plt.barh(iris.feature_names, feature_importances)
plt.xlabel("Feature Importance")
plt.title("Feature Importance in Decision Tree Classifier")
plt.show()
```

In prepruning hyperparameters are optimized during cross-validation so in this case Pre-pruning : GridSearchCV found optimal parameters and that allowed complex tree more splits mean more features are used in the decision making process
