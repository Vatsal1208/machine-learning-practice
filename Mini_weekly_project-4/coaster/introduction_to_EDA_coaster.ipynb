```python
duplicated_rows = df.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"])
```

```python
rows_to_keep = ~duplicated_rows
```

```python
rows_to_keep.sum()
```

```python
df = df.loc[rows_to_keep].reset_index(drop=True).copy()
```

```python
df.head()
```

~ Flips True to False and vice versa So now it is True only for the first occurrence of each unique (Coaster_Name, Location, Opening_Date) and False for later duplicates.

so to keep only one row per unique coaster (name+location + opening date) drop duplicate rows for the same coaster 

[pandas opposite of drop duplicates using bitwise ~ ](https://stackoverflow.com/questions/55121831/pandas-opposite-of-drop-duplicates-keep-first)

```python
Q1 = df["Speed_mph"].quantile(0.25)
Q3 = df["Speed_mph"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Outliers: {len(df[(df['Speed_mph'] < lower) | (df['Speed_mph'] > upper)])}")
print(f"Lower bound: {lower}, Upper bound: {upper}")
print(f"IQR: {IQR}")
print(f"25% range: {Q1} and 75% range: {Q3}")
```

```python
lower_outliers = df["Speed_mph"] < lower
lower_outliers.value_counts()
```

```python
upper_outliers = df["Speed_mph"] > upper
upper_outliers.value_counts()
```

```python
df[df["Speed_mph"] > upper].count()
```

```python
df[df["Speed_mph"] < lower].count()
```

```python
new_df = df[(df["Speed_mph"] > float(lower)) & (df["Speed_mph"] < float(upper))]
new_df = pd.DataFrame(new_df)
```
